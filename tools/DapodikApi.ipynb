```python
import requests

# konfigurasi url api
url = 'http://dapo.dikdasmen.kemdikbud.go.id/rekap'
url_datasekolah = '{}/dataSekolah?id_level_wilayah={}&kode_wilayah={}&semester_id={}'
url_progresSP = '{}/progresSP?id_level_wilayah={}&kode_wilayah={}&semester_id={}'
url_detailsekolah = '{}/sekolahDetail?semester_id={}&sekolah_id={}'

# definisi semester dan kode kabupaten
# 191100 untuk kabupaten bulukumba
semester = 20162
kode_kabupaten = 191100
```

```python
# mengunduh semua kecamatan per kode kabupaten
r = requests.get(url_datasekolah.format(url, 2, kode_kabupaten, semester).replace(" ", ""))
# data_kecamatan akan bernilai list / array untuk semua daftar informasi kecamatan
data_kecamatan = r.json()
```

```python
from prettytable import PrettyTable

x = PrettyTable(["status_sekolah", "kecamatan", "nama_sekolah", "pd"])

# mengunduh data sekolah untuk semua kecamatan
for kecamatan in data_kecamatan:
    # variable kecamatan bernilai object
    # lakukan print(kecamatan.keys()) untuk mengetahui key yang tersedia
    print('Mengunduh detail daftar sekolah {}'.format(kecamatan.get('nama')))
    # kode kecamatan terdapat pada kata kunci kode wilayah
    kode_kecamatan = kecamatan.get('kode_wilayah')

    # mengunduh data sekolah per kecamatan
    r = requests.get(url_progresSP.format(url, 3, kode_kecamatan, semester).replace(" ", ""))
    # daftar_sekolah akan bernilai list / array 
    daftar_sekolah = r.json()
    
    for sekolah in daftar_sekolah:
        # variable sekolah merupakan object / dictionary untuk data per sekolah
        # lakukan print(sekolah.keys()) untuk mengetahui key yang tersedia
        nama_sekolah = sekolah.get('nama')
        status_sekolah = sekolah.get('status_sekolah')
        kecamatan = sekolah.get('induk_kecamatan')
        pd = sekolah.get('pd')
        if pd is None or pd == 'None':
            pd = 0
        x.add_row([status_sekolah, kecamatan, nama_sekolah, pd])
print(x)
```
